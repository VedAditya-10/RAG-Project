```python
import os
import requests
import json
from dotenv import load_dotenv
import chromadb

# Load environment variables
load_dotenv()

# ChromaDB setup
CHROMA_PATH = "chroma_db"
chroma_client = chromadb.PersistentClient(path=CHROMA_PATH)
collection = chroma_client.get_or_create_collection(name="Syllabus")

def query_ollama(prompt, model="llama3.2"):
    """Query Ollama API running locally"""
    try:
        response = requests.post('http://localhost:11434/api/generate',
                               json={
                                   'model': model,
                                   'prompt': prompt,
                                   'stream': False,
                                   'options': {
                                       'temperature': 0.2,
                                       'num_predict': 500
                                   }
                               })
        
        if response.status_code == 200:
            return response.json()['response']
        else:
            return f"Error: {response.status_code} - {response.text}"
    
    except requests.exceptions.ConnectionError:
        return "Error: Cannot connect to Ollama. Make sure Ollama is running (try 'ollama serve' in terminal)"
    except Exception as e:
        return f"Error: {str(e)}"

# Ask user for query
user_query = input("What do you want to know about the syllabus?\n\n")

# Retrieve relevant documents from ChromaDB
results = collection.query(
    query_texts=[user_query],
    n_results=4
)

print("Retrieved docs:", results['documents'])

# Construct system prompt
system_prompt = f"""You are a helpful assistant. You answer questions about the Syllabus of the Computer Science Engineering Department.
Answer only based on the data provided below. Do not use your internal knowledge and do not make things up.
If you don't know the answer, say: I don't know.

Data:
{results['documents']}

Question: {user_query}

Answer:"""

# Generate response using Ollama
print("Generating response with Ollama...")
response_text = query_ollama(system_prompt, model="llama3.2")

print("\n\n---------------------\n\n")
print(response_text)
```
